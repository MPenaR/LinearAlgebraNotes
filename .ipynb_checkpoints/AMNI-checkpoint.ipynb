# Número de condición:

Queremos estudiar la dependencia de la solución $\mathbf{x}$ del sistema compatible determinado:
$$\mathbf{A}\mathbf{x}=\mathbf{b}$$
con respecto a una variación pequeña en el término independiente $\Delta\mathbf{b}$. Para ello planteamos el sistema perturbado 
$$\mathbf{A}\mathbf{y}=\mathbf{b}+\Delta\mathbf{b}$$
cuya solución existe y es única poEs decir, si llamamos $\Delta\mathbf{x}$ a la diferencia entre la solución del sistema, perturbado y la solución del sistema original:
$$
\mathbf{A}\left(\mathbf{x}+\Delta\mathbf{x}\right) = \mathbf{b} + \Delta \mathbf{b}
$$

Acotando se tiene que:
$$
\frac{\Vert \Delta\mathbf{x}\Vert}{\mathbf{\Vert x\Vert}}
$$
(falta la demostración pero si no os vais a ir a dormir)

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
```

Sean dos sistemas $\mathbf{A}\mathbf{x}=\mathbf{b}$ y $\mathbf{B}\mathbf{x}=\mathbf{b}$  con
$$
\mathbf{A}=\begin{pmatrix}
10 & 1 \\
1 & 10
\end{pmatrix},\quad
\mathbf{B}=\begin{pmatrix}
10 & 9 \\
9 & 10
\end{pmatrix}\,\quad
\mathbf{b}=\begin{pmatrix}
1 \\
1
\end{pmatrix}.
$$

El número de condición de la matriz $\mathbf{A}$ es $\mathrm{cond}\left(\mathbf{A}\right) = \frac{11}{9}\approx1.22$ mientras que el número de condición de la matriz $\mathbf{B}$ es $\mathrm{cond}\left(\mathbf{B}\right) = 19$. Esto se debe a que las filas de la matriz $\mathbf{B}$ son prácticamente paralelas mientras que las de la matriz $\mathbf{A}$ no. Si llamamos $r_1$ y $r_2$ a las rectas correspondientes a cada una de las ecuaciones del primer sistema:
$$r_1 = \left\lbrace 10x + 1y = 1 \right\rbrace$$
$$r_2 = \left\lbrace 1x + 10y = 1 \right\rbrace$$
y $s_1$ y $s_2$ a las del segundo:
$$s_1 = \left\lbrace 10x + 9y = 1 \right\rbrace$$
$$s_2 = \left\lbrace 9x + 10y = 1 \right\rbrace$$
podemos representarlas gráficamente, junto con la solución del sistema (única en ambos casos). 

```python
fig, ax = plt.subplots(nrows=1,ncols=2, figsize = [ 12,6])
A = np.array([
    [10,1],
    [1,10]])
b = np.array([1,1])
x = np.linalg.inv(A).dot(b)
ax[0].plot([0,1/10],[1,0])
ax[0].plot([0,1],[1/10,0])
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$y$',rotation=0)
ax[0].legend(['$r_1$','$r_2$'])
ax[0].set_title('$\\mathrm{{cond}}\\left(\\mathbf{{A}}\\right) = 1.22\dots$')
ax[0].plot(x[0],x[1],'ok')
#e = 0.05
#ax.plot([0,(1+e)/10],[(1+e),0],'--k')
#ax.plot([0,(1+e)],[(1+e)/10,0],'--k')
#ax.plot([0,(1-e)/10],[(1-e),0],'--k')
#ax.plot([0,(1-e)],[(1-e)/10,0],'--k')
ax[0].axis('square')
ax[0].set_xlim([0,0.2])
ax[0].set_ylim([0,0.2])

#x1 = np.linalg.inv(A).dot(b + np.array([e,e]))
#x2 = np.linalg.inv(A).dot(b + np.array([e,-e]))
#x3 = np.linalg.inv(A).dot(b + np.array([-e,-e]))
#x4 = np.linalg.inv(A).dot(b + np.array([-e,e]))
#ax.add_patch(Polygon(xy=[x1,x2,x3,x4],))
B = np.array([
    [10,9],
    [9,10]])
b = np.array([1,1])
y = np.linalg.inv(B).dot(b)
ax[1].plot([0,1/10],[1/9,0])
ax[1].plot([0,1/9],[1/10,0])
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$y$',rotation=0)
ax[1].legend(['$s_1$','$s_2$'])
ax[1].set_title('$\\mathrm{{cond}}\\left(\\mathbf{{B}}\\right) = 19$')
ax[1].plot(y[0],y[1],'ok')
#e = 0.05
#ax.plot([0,(1+e)/10],[(1+e)/9,0],'--k')
#ax.plot([0,(1+e)/9],[(1+e)/10,0],'--k')
#ax.plot([0,(1-e)/10],[(1-e)/9,0],'--k')
#ax.plot([0,(1-e)/9],[(1-e)/10,0],'--k')
ax[1].axis('square')
ax[1].set_xlim([0,0.2])
ax[1].set_ylim([0,0.2])

#y1 = np.linalg.inv(B).dot(b + np.array([e,e]))
#y2 = np.linalg.inv(B).dot(b + np.array([e,-e]))
#y3 = np.linalg.inv(B).dot(b + np.array([-e,-e]))
#y4 = np.linalg.inv(B).dot(b + np.array([-e,e]))
#ax.add_patch(Polygon(xy=[y1,y2,y3,y4],))
plt.show()
```

Sin embargo, si modificamos el vector $\mathbf{b}$ una cantidad pequeña $\Delta\mathbf{b}$ las rectas anteriores se desplazan paralelamente. En la figura siguiente se muestran las rectas correspondientes a 
$$
\Delta\mathbf{b} = \begin{pmatrix} \pm 0.05 \\ \pm 0.05\end{pmatrix} 
$$

Como puede comprobarse, en el sistema correspondiente a la matriz $\mathbf{A}$ la nueva solución queda confinada en una región de tamaño pequeño (marcada en azul) mientras que esta región para el sistema de la matriz $\mathbf{B}$ es mucho mayor.

```python
fig, ax = plt.subplots(nrows=1,ncols=2, figsize = [ 12,6])
A = np.array([
    [10,1],
    [1,10]])
b = np.array([1,1])
x = np.linalg.inv(A).dot(b)
ax[0].plot([0,1/10],[1,0])
ax[0].plot([0,1],[1/10,0])
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$y$',rotation=0)
ax[0].legend(['$r_1$','$r_2$'])
ax[0].set_title('$\\mathrm{{cond}}\\left(\\mathbf{{A}}\\right) = 1.22\dots$')
ax[0].plot(x[0],x[1],'ok')
e = 0.05
ax[0].plot([0,(1+e)/10],[(1+e),0],'--k')
ax[0].plot([0,(1+e)],[(1+e)/10,0],'--k')
ax[0].plot([0,(1-e)/10],[(1-e),0],'--k')
ax[0].plot([0,(1-e)],[(1-e)/10,0],'--k')
ax[0].axis('square')
ax[0].set_xlim([0,0.2])
ax[0].set_ylim([0,0.2])

x1 = np.linalg.inv(A).dot(b + np.array([e,e]))
x2 = np.linalg.inv(A).dot(b + np.array([e,-e]))
x3 = np.linalg.inv(A).dot(b + np.array([-e,-e]))
x4 = np.linalg.inv(A).dot(b + np.array([-e,e]))
ax[0].add_patch(Polygon(xy=[x1,x2,x3,x4],))
B = np.array([
    [10,9],
    [9,10]])
b = np.array([1,1])
y = np.linalg.inv(B).dot(b)
ax[1].plot([0,1/10],[1/9,0])
ax[1].plot([0,1/9],[1/10,0])
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$y$',rotation=0)
ax[1].legend(['$s_1$','$s_2$'])
ax[1].set_title('$\\mathrm{{cond}}\\left(\\mathbf{{B}}\\right) = 19$')
ax[1].plot(y[0],y[1],'ok')
ax[1].plot([0,(1+e)/10],[(1+e)/9,0],'--k')
ax[1].plot([0,(1+e)/9],[(1+e)/10,0],'--k')
ax[1].plot([0,(1-e)/10],[(1-e)/9,0],'--k')
ax[1].plot([0,(1-e)/9],[(1-e)/10,0],'--k')
ax[1].axis('square')
ax[1].set_xlim([0,0.2])
ax[1].set_ylim([0,0.2])

y1 = np.linalg.inv(B).dot(b + np.array([e,e]))
y2 = np.linalg.inv(B).dot(b + np.array([e,-e]))
y3 = np.linalg.inv(B).dot(b + np.array([-e,-e]))
y4 = np.linalg.inv(B).dot(b + np.array([-e,e]))
ax[1].add_patch(Polygon(xy=[y1,y2,y3,y4],))
plt.show()
```
